#### Interpretation of Skewness
1. Positive Skewness: If the skewness values (calculated from the data) are positive, it implies the distribution is right-skewed, meaning:
    - Most data points are clustered toward lower values.
    - The tail is longer on the higher side.
    - In the context of EV charging infrastructure, this may suggest a large proportion of locations with fewer sanctioned chargers or stations, while a few locations have significantly higher counts.
2. Negative Skewness: If skewness values are negative, the distribution is left-skewed, indicating:
    - Most data points are clustered toward higher values.
    - The tail is longer on the lower side.
    - This might indicate that a majority of locations have many chargers or stations, with only a few having fewer.
3. Near-Zero Skewness:  A skewness close to zero would imply a symmetric distribution, where the facilities are evenly distributed across locations.

## Bivariate Analysis:

```python
# Sort the DataFrame by the values in descending order
df1_sorted = df1.sort_values(by='No. of Electric Vehicle (EV) Chargers Sanctioned', ascending=False)

# Bar Graph with sorted values
plt.figure(figsize=(15, 10))
sns.barplot(
    x='State/UT', 
    y='No. of Electric Vehicle (EV) Chargers Sanctioned', 
    data=df1_sorted,  # Use the sorted DataFrame
    palette='viridis'
)
plt.title('No. of EV Chargers Sanctioned by State/UT (Sorted)', fontsize=16)
plt.xticks(rotation=45)
plt.xlabel('State/UT',  fontsize=14)
plt.ylabel('No. of EV Chargers Sanctioned',  fontsize=14)
plt.tight_layout()
plt.show()
```

### Inference from the Bar Graph:
1. Dominance of One State/Region: A single state or region, Maharashtra and Tamil nadu has significantly higher EV chargers sanctioned compared to others.
2. Large Gap: The gap between the top state and the rest is substantial, indicating a concentrated focus or disparity in EV charger deployment across states/UTs.
3. Moderate Distribution: A handful of states (Gujarat, Andhra Pradesh, etc.) have moderate numbers of EV chargers sanctioned, forming a second tier.
4. Low Representation: Many states/UTs have minimal chargers sanctioned, reflecting either low adoption, policy prioritization, or infrastructural challenges in these regions.

### Pie Chart:

```python
# Pie Chart for No. of EV Chargers Sanctioned
plt.figure(figsize=(15, 15))
plt.pie(
    df1_sorted['No. of Electric Vehicle (EV) Chargers Sanctioned'], 
    labels=df1_sorted['State/UT'], 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=sns.color_palette('coolwarm', len(df1_sorted))
)
plt.title('Proportion of EV Chargers Sanctioned by State/UT', fontsize=16)
plt.tight_layout()
plt.show()
```

### Inferernce based on the Pie chart :

- Maharashtra has the highest number of sanctioned charging stations that 5.5% of the total follwed by Tamil Nadu with 4.9% and then Gujrat with 4.8%. This shows people in these states are more inclinded towards buying EVs and government is sanctioning the stations.

- The eastern states like Assaam, Meghalaya and Bihar still lags when it comes to sanctioned stations.
